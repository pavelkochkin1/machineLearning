```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', palette='Dark2')
```

```python
inp = load_iris()
data = pd.DataFrame(data= np.c_[inp['data'], inp['target']],
                     columns= inp['feature_names'] + ['target'])
```

```python
data.head()
```

```python
data.info()
```

```python
sns.pairplot(data, hue='target', palette='Dark2')
None
```

```python
sns.histplot(data.target)
None
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('target', axis=1), data.target, test_size=0.2, random_state=42)
```

```python
sns.pairplot(X_test, diag_kind='kde')
None
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)

pred = model.predict(X_test)

accuracy_score(pred, y_test)
```

```python
X_test['class'] = pred

sns.pairplot(X_test, hue='class', palette='Dark2')
```

```python
X = pd.read_csv('weight-height.csv')

X.Weight /= 2.20462 # в кг
X.Height *= 2.54 # в метры

X.head()
```

```python
print(X.info())
print(X.describe())
```

```python
sns.pairplot(X, hue='Gender')
None
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1, 2, 1)
sns.histplot(X.Height)
plt.subplot(1, 2, 2)
sns.histplot(X.Weight)
```

```python
from sklearn.preprocessing import LabelBinarizer

preproc = LabelBinarizer()
X.Gender = preproc.fit_transform(X.Gender)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.drop('Height', axis=1), X.Height, test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)

mean_squared_error(pred, y_test)
```
