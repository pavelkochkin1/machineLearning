```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(columns=['Alley', 'PoolQC', 'Fence', 'MiscFeature'], axis=1, inplace = True)
data
```

```python
data.LotFrontage.fillna(70, inplace = True)
```

```python
data.FireplaceQu.value_counts()
```

```python
data.FireplaceQu = data.FireplaceQu.map({'Gd':1, 'TA':2, 'Fa':3, 'Ex': 4, 'Po': 5})
```

```python
data.Fireplaces.value_counts()
```

```python
data.FireplaceQu.value_counts()
```

```python
data.groupby('FireplaceQu').SalePrice.mean()
```

```python
data[data.Fireplaces == 0].FireplaceQu.isna().value_counts()
```

```python
data.FireplaceQu.fillna(0, inplace=True)
```

```python
data.GarageType.fillna('No Garage', inplace=True)
data.GarageFinish.fillna('No Garage', inplace=True)
data.GarageQual.fillna('No Garage', inplace=True)
data.GarageYrBlt.fillna(0 ,inplace=True)
data.GarageCond.fillna('No Garage', inplace=True)
```

```python
data.info()
```

```python
data.dropna(axis=0, how='any', inplace=True)
data.SalePrice = round(data.SalePrice / 1000)
data.SalePrice
```

```python
data = pd.get_dummies(data)
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('SalePrice', axis=1), data.SalePrice, test_size=0.33)
```

```python
%%time
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import  mean_absolute_error
model_linear = RandomForestRegressor(max_depth=10, n_estimators=300, min_samples_leaf=10)
model_linear.fit(X_train, y_train)
```

```python
print('Train MSE: ', mean_absolute_error(y_train, model_linear.predict(X_train)))
print('Test MSE: ', mean_absolute_error(y_test, model_linear.predict(X_test)))
```
