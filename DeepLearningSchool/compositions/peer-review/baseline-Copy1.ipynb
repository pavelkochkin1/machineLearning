```python
# YOUR CODE
```

```python
def show_roc_auc(y, y_pred, name, y2=None, y_pred2=None, name2=None):
# ф-ция строит график roc_auc
    fpr, tpr, _ = roc_curve(y, y_pred)
    pred_auc_score = roc_auc_score(y, y_pred) 
    plt.plot(fpr, tpr, label=f'{name}: auc={round(pred_auc_score, 4)}')
    
    if isinstance(name2,str):
        fpr, tpr, _ = roc_curve(y2, y_pred2)
        pred_auc_score = roc_auc_score(y2, y_pred2) 
        plt.plot(fpr, tpr, label=f'{name2}: auc={round(pred_auc_score, 4)}')           
         
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.title('ROC curve')
    plt.legend(loc='best')
    plt.show()
    return
```

```python
# логистическая регрессия (LogisticRegression): перебор пораметра С (вручную):
C_grid= [100, 10, 1, 0.1, 0.01, 0.001]
plt.figure(1)
for C in C_grid:
    logreg = LogisticRegression(random_state=0, C=C).fit(X_train, y_train)
#     logreg.predict_proba(X_test)
#     logreg.score(X_test, y_test)
    y_test_pred = logreg.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_test_pred)
    pred_auc_score = roc_auc_score(y_test, y_test_pred)

    
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot(fpr, tpr, label=f'LogReg, C={C}, auc={round(pred_auc_score, 4)}')

plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
```

```python
# логистическая регрессия (LogisticRegressionCV): с кроссвалидацией

logregCV = LogisticRegressionCV(cv=5, random_state=0, scoring='roc_auc', n_jobs=-1, refit=True)
logregCV.fit(X_train, y_train)

y_pred = logregCV.predict_proba(X)[:, 1]

show_roc_auc(y_train, logregCV.predict_proba(X_train)[:, 1], f'LogRegCV: train',
             y_test, logregCV.predict_proba(X_test)[:, 1], f'LogRegCV: test')
```

```python
C_grid= [100, 10, 1, 0.1, 0.01, 0.001]

CV_model = GridSearchCV(estimator=LogisticRegression(),                        
                            param_grid={'C': C_grid},
                            cv=5, 
                            scoring='roc_auc',  ## метрика, f1 - мера для бинарной классификации, чем ближе к 1, тем лучше
                            n_jobs=-1, 
                            verbose=10)
CV_model.fit(X_train, y_train)
means = CV_model.cv_results_['mean_test_score'] #
error = CV_model.cv_results_['std_test_score']
    
plt.figure(figsize=(15,8))
plt.title('choose C')
plt.xscale('log')
plt.plot(C_grid, means, label='mean values of score', color='red', lw=3)
plt.fill_between(C_grid, means - 2 * error, means + 2 * error, 
                         color='green', label='filled area between errors', alpha=0.5)
legend_box = plt.legend(framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.xlabel('parameter')
plt.ylabel('roc_auc')
plt.show()
```
